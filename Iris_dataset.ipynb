```
import pandas as pd
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
import seaborn as sns
```

```
url='http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```
urllib.request.urlretrieve(url)
```

```
df=pd.read_csv(url,sep=',')
```

```
field_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
```

```
df.columns=field_names
```

```
#the data sets are divided according to class
df1=df.iloc[0:49, : ]
df2=df.iloc[50:99, : ]
df3=df.iloc[100:149, : ]
```

```
df['class'].value_counts()
```

```
cols=df.columns
print(cols)
```

```
#adding columns total_petal and total_sepal
cols=df.columns
cols1=cols[2:4]
cols2=cols[0:2]
cols3=cols[0:4]
c=df[cols3]
a=df[cols1]
b=df[cols2]
df['total_petal']=a[cols1].sum(axis=1)
df['total_sepal']=b[cols2].sum(axis=1)
df['total_value']=c[cols3].sum(axis=1)
print(df)
```

```
#finding the number of flowers greater than average petal lenght
mean_petal_lenght=np.mean(df['petal_length'])
greater_petal=df.where(df['petal_length'] > mean_petal_lenght)
print("{} flowers have greater petal lenth than the average petal length".format(greater_petal['petal_length'].count()))
```

```
mean_sepal_lenght=np.mean(df['sepal_length'])
greater_sepal=df.where(df['sepal_length'] > mean_sepal_lenght)
print("{} flowers have greater sepal length than the average sepal length".format(greater_sepal['sepal_length'].count()))
```

```
mean_petal_width=df['petal_width'].mean()
greater_pw=df.where(df['petal_width']>mean_petal_width)
print("{} flowers having greater petal width than average petal width".format(greater_pw['petal_width'].count()))
```

```
mean_sepal_width=df['sepal_width'].mean()
greater_sw=df.where(df['sepal_width']>mean_sepal_width)
print("{} flowers having greater sepal width than average sepal width".format(greater_sw['sepal_width'].count()))
```
